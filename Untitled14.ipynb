```python
# Studentâ€“Course Management System using sets

students = {}  # Dictionary to store student name and their courses

def add_student():
    name = input("Enter student name: ")
    courses = input("Enter courses (comma-separated): ").split(',')
    students[name] = set(course.strip() for course in courses)
    print("Student added successfully.\n")

def display_students():
    if not students:
        print("No students found.\n")
    else:
        for name, courses in students.items():
            print(f"{name}: {courses}")
        print()

def get_course_students(course):
    return {name for name, courses in students.items() if course in courses}

def intersection_courses():
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    result = get_course_students(c1) & get_course_students(c2)
    print(f"Students in both {c1} and {c2}: {result}\n")

def union_courses():
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    result = get_course_students(c1) | get_course_students(c2)
    print(f"Students in either {c1} or {c2}: {result}\n")

def difference_courses(first=True):
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    if first:
        result = get_course_students(c1) - get_course_students(c2)
        print(f"Students only in {c1} but not in {c2}: {result}\n")
    else:
        result = get_course_students(c2) - get_course_students(c1)
        print(f"Students only in {c2} but not in {c1}: {result}\n")

def symmetric_difference_courses():
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    result = get_course_students(c1) ^ get_course_students(c2)
    print(f"Students exclusively in one of {c1} or {c2}: {result}\n")

def subset_superset_check():
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    s1 = get_course_students(c1)
    s2 = get_course_students(c2)
    print(f"{c1} is subset of {c2}: {s1.issubset(s2)}")
    print(f"{c1} is superset of {c2}: {s1.issuperset(s2)}\n")

def disjoint_check():
    c1 = input("Enter first course: ")
    c2 = input("Enter second course: ")
    s1 = get_course_students(c1)
    s2 = get_course_students(c2)
    print(f"{c1} and {c2} are disjoint: {s1.isdisjoint(s2)}\n")

def menu():
    while True:
        print("1. Add Student")
        print("2. Display Students")
        print("3. Students in both courses (Intersection)")
        print("4. Students in either course (Union)")
        print("5. Students only in first course (Difference)")
        print("6. Students only in second course (Difference)")
        print("7. Students exclusively in one course (Symmetric Difference)")
        print("8. Check Subset/Superset")
        print("9. Check Disjoint")
        print("10. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_student()
        elif choice == '2':
            display_students()
        elif choice == '3':
            intersection_courses()
        elif choice == '4':
            union_courses()
        elif choice == '5':
            difference_courses(first=True)
        elif choice == '6':
            difference_courses(first=False)
        elif choice == '7':
            symmetric_difference_courses()
        elif choice == '8':
            subset_superset_check()
        elif choice == '9':
            disjoint_check()
        elif choice == '10':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Try again.\n")

menu()
```
